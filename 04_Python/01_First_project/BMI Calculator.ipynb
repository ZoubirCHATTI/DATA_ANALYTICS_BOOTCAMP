**BMI Calculator**
This project demonstrates three different ways to create a BMI (Body Mass Index) calculator in Python.
The program uses a personâ€™s weight (in pounds) and height (in inches) to compute their BMI and determine their weight category.

*Method 1*: All steps (input, calculation, and classification) are written directly in the main code without using functions.

*Method 2*: A single function is used to handle the entire process, making the code cleaner and easier to reuse.

*Method 3*: A more structured approach where a function receives weight, height, and name as parameters, computes the BMI, and prints the result.

These three methods show how the same task can be done in Python from a simple script to a more organized and modular code structure.

```python
weight= int(input("Enter your weight in pounds: "))

height= int(input("Enter your height in inches: "))

BMI =(weight*703)/(height * height) 

name=input("enter your name: ")
print(BMI)

if BMI > 0:
    if BMI < 18.5:
        print(name+", You are underweight")
    elif (BMI <24.9):
        print(name+",You are normal")
    elif (BMI <29.9):
        print(name+",You are overweight")
    elif (BMI <34.9):
        print(name+",You are obese")
    elif (BMI <39.9):
        print(name+",You are severely obese")    
    else:
        print(name+",You are morbidely obese")    
```
