**World Population Exploratory Data Analysis (EDA)**

This project performs an exploratory data analysis on a global population dataset covering years from 1970 to 2022. The goal is to identify growth trends, density patterns, and demographic distribution across different continents.

```python
#Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Data loading
df=pd.read_csv("https://raw.githubusercontent.com/ZoubirCHATTI/DATA_ANALYTICS_BOOTCAMP/main/04_Python/03_Exploratory_Data_Analysis/data/world_population.csv")
df
```

```python
pd.set_option("display.float_format", lambda x: "%.2f" % x)
```

```python
#Checking data types and missing valuees
df.info()
```

```python
#Statistical summary of numerical columns 
df.describe()
```

```python
#Checking for null values across features
df.isnull().sum()
```

```python
#Checking unique value for each feature
df.nunique()
```

```python
#Sorting values world population percentage
df.sort_values(by="World Population Percentage", ascending = False).head(10)
```

```python
#Correlation of different numerical parameters
df.corr(numeric_only=True)
```

```python
#Creation of a heatmap to better correlation between different parameters
plt.figure(figsize=(8, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

```python
pd.set_option("display.max.rows", 250)
```

```python
#Grouping by Continent
df.groupby("Continent").mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
```

```python
# Aggregating population by continent and sorting by 2022 population
df2=df.groupby("Continent")[df.columns[5:13]].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
```

```python
df3=df2.transpose()
```

```python
# Visualizing population growth trends over time (1970 - 2022) for different continents
df3.plot(kind="line", figsize=(12, 5))
plt.xticks(rotation=45)
plt.gca().invert_xaxis()
plt.grid(True)
plt.legend()
```

```python
#Boxplots to detect outliers in population distribution
df.boxplot(column=["1970 Population","1980 Population","1990 Population","2000 Population","2010 Population","2020 Population","2022 Population"],figsize=(12, 6))
plt.xticks(rotation=90)
plt.legend()
```
